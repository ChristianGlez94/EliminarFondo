```python
# Instalar las bibliotecas necesarias
!pip install rembg pillow onnxruntime

from rembg import remove
from PIL import Image
from google.colab import files
import io

# Subir la imagen original
print("Sube la imagen original para eliminar el fondo")
uploaded = files.upload()

# Procesar la imagen
for filename in uploaded.keys():
    # Abrir la imagen subida por el usuario
    input_image = Image.open(io.BytesIO(uploaded[filename]))
    input_image = input_image.convert("RGBA")  # Convertir a RGBA para asegurar el canal alfa

    print("Procesando la imagen para eliminar el fondo...")

    # Convertir la imagen a bytes y eliminar el fondo
    with io.BytesIO() as img_byte_arr:
        input_image.save(img_byte_arr, format="PNG")
        img_byte_arr = img_byte_arr.getvalue()  # Obtener los bytes de la imagen
        output_image_bytes = remove(img_byte_arr)  # Eliminar fondo

    # Cargar la imagen resultante en un objeto PIL
    output_image = Image.open(io.BytesIO(output_image_bytes))
    output_image = output_image.convert("RGBA")  # Asegurarse del formato RGBA

    # Guardar la imagen sin fondo
    output_image.save("sin_fondo.png", format="PNG")
    print("Imagen sin fondo guardada como 'sin_fondo.png'")

# Subir el nuevo fondo
print("Sube el nuevo fondo que deseas agregar")
uploaded_background = files.upload()

# Leer y procesar el fondo
for bg_filename in uploaded_background.keys():
    background_image = Image.open(io.BytesIO(uploaded_background[bg_filename]))
    background_image = background_image.convert("RGBA")

    # Asegurarse de que el fondo tenga el mismo tamaño que la imagen sin fondo
    background_image = background_image.resize(output_image.size)

    # Combinar la imagen sin fondo con el nuevo fondo
    combined_image = Image.alpha_composite(background_image, output_image)

    # Guardar la imagen combinada
    combined_image.save("con_nuevo_fondo.png", format="PNG")
    print("Imagen con nuevo fondo guardada como 'con_nuevo_fondo.png'")

# Descargar las imágenes procesadas
print("Descarga las imágenes procesadas")
files.download("sin_fondo.png")
files.download("con_nuevo_fondo.png")
```
